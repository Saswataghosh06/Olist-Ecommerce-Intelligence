# Ingesting or loading CSV File to duck DB

import duckdb
import os

# 1. We define a persistent database file named 'olist_warehouse.db' inside the data folder
db_path = "../data/olist_warehouse.db"
con = duckdb.connect(db_path)

print(f"Connected to persistent database at: {db_path}\n")

raw_dir = "../data/raw/"
files = [f for f in os.listdir(raw_dir) if f.endswith('.csv')]

# 2. Loop through and permanently ingest the data
for file in files:
    table_name = file.replace('.csv', '')
    file_path = os.path.join(raw_dir, file)
    
    print(f"Ingesting {file} into table '{table_name}'...")
    
    # This creates the table inside olist_warehouse.db permanently
    con.execute(f"""
        CREATE OR REPLACE TABLE {table_name} AS 
        SELECT * FROM read_csv_auto('{file_path}');
    """)

print("\nIngestion complete! Your Bronze Layer is ready.")
con.close()
